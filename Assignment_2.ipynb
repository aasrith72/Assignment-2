```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
import textwrap
import traceback

tokenizer = None
model = None
try:
    print("\nAttempting to load GPT2Tokenizer and GPT2LMHeadModel...")
    tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
    model = GPT2LMHeadModel.from_pretrained("gpt2")
    print("GPT-2 Tokenizer and Model loaded successfully.")

    tokenizer.pad_token_id = tokenizer.eos_token_id
    model.config.pad_token_id = model.config.eos_token_id # Also set for the model's config
    print(f"Pad token ID set to tokenizer's EOS token ID: {tokenizer.pad_token_id}")

except Exception as e:
    print(f"Failed to load GPT2Tokenizer or GPT2LMHeadModel: {e}")
    print("Story generation function may not work.")
    tokenizer = None
    model = None

def generate_story(prompt: str, temperature: float = 0.7) -> str:
    """
    Generates a story continuation using GPT-2 based on a given prompt and temperature.

    Args:
        prompt (str): The starting text for the story.
        temperature (float): Controls the randomness of the generation.
                             Higher values (e.g., 1.0) make the output more random/creative.
                             Lower values (e.g., 0.3) make it more deterministic/focused.
                             Default is 0.7.

    Returns:
        str: The generated story continuation, or an error message if generation fails.
    """
    if tokenizer is None or model is None:
        return "Error: GPT-2 tokenizer or model not loaded. Cannot generate story."

    try:
        input_ids = tokenizer.encode(prompt, return_tensors='pt')

        output_sequences = model.generate(
            input_ids=input_ids,
            max_length=len(input_ids[0]) + 100,
            do_sample=True,
            temperature=temperature,
            top_k=50,
            top_p=0.95,
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id
        )

        generated_text = tokenizer.decode(output_sequences[0], skip_special_tokens=True)
        continuation = generated_text[len(prompt):].strip()
        return continuation

    except Exception as e:
        return f"An error occurred during story generation: {e}\n{traceback.format_exc()}"

# --- Test Cases ---
if __name__ == "__main__":
    if tokenizer and model: # Only run tests if model and tokenizer loaded successfully
        print("\n--- Testing Story Generation with Different Temperatures ---")
        base_prompt = "Once upon a time, in a land far, far away,"

        temperature_values = [0.3, 0.7, 1.0, 1.3] # Test with different temperature values [cite: 3]

        for temp in temperature_values:
            print(f"\nPrompt: \"{base_prompt}\"")
            print(f"Temperature: {temp}")
            generated_story = generate_story(base_prompt, temperature=temp)
            print("Generated Story:")
            print(textwrap.fill(generated_story, width=80)) # Observe and record how the generation changes [cite: 3]
            print("-" * 80)
    else:
        print("\nSkipping story generation tests because GPT-2 model or tokenizer failed to load.")
```
